```python
def remove_usernames_links(tweet):

    tweet = re.sub('@[^\s]+','',tweet)
    tweet = re.sub('http[^\s]+','',tweet)
    tweet = re.sub('#[^\s]+','',tweet)
    tweet = re.sub('[ههه]+','',tweet)

    return tweet
```

```python
def remove_nonalphabetical_characters(tweet):

  l=[]
  for i in tweet:
    if (i.isalpha() and (i not in string.ascii_letters)) or i == ' ':
      l.append(i)
    else:
      l.append(' ')
  new_tweet=''.join(l)

  return new_tweet
```

```python
df['modified_tweet'] = df['text'].apply(remove_usernames_links)
df['new_modified_tweet']=df['modified_tweet'].apply(remove_nonalphabetical_characters)
```

```python
from nltk.stem import WordNetLemmatizer
#defining the object for Lemmatization
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
nltk.download('wordnet')
```

```python
#defining the function for lemmatization
def lemmatizer(text):
  lemm_text = [wordnet_lemmatizer.lemmatize(word) for word in text]
  return lemm_text
df['tweet_lemmatized']=df['no_stopwords'].apply(lemmatizer)
```

```python
#importing the Stemming function from nltk library
from nltk.stem.arlstem import ARLSTem
#defining the object for stemming
stemmer = ARLSTem()
```
